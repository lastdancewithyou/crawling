```python
def process_urls(self, urls, session):
    # 최대 리디렉션 횟수 5
    max_redirects = 5

    for url in urls:
        total_word_count = 0

        try:
            res = session.get('http://'+url, timeout=3)




            self.logger.info(f'{url}: success connect.')

            self.connected_url_count += 1
            # BeautifulSoup 객체 생성
            soup = BeautifulSoup(res.text, "html.parser")  

            # HTML 소스코드에서 텍스트만 추출하기
            inner_text = soup.get_text()
            # inner_text = res.text
            for word in self.keyword_list:
                word_count = inner_text.count(word)
                total_word_count += word_count

            # 리디렉션 횟수가 제한을 넘어서면 예외를 발생시킴
            if res.is_redirect and len(res.history) >= max_redirects:
                raise TooManyRedirects()

        # 타임 아웃일 경우
        except Timeout:
            self.logger.warning(f'{url}: The request timed out.')

        # 인터넷 연결이 안된 경우
        except ConnectionError:
            self.logger.warning(f'{url}: not internet connection.')

        # 서버에서 요청에 대한 HTTP 응답 코드가 400 이상인 경우
        except HTTPError:
            self.logger.warning(f'{url}: HTTPError.')

        # 리디렉션 횟수가 제한을 넘어선 경우
        except TooManyRedirects:
            self.logger.warning(f'{url}: Too many redirects.')


        # gzip으로 설정된 Content-Encoding 처리
        except ContentDecodingError as e:
            self.logger.warning(f"{url}: ContentDecodingError occurred: {e}")

        except Exception as e:
                # 예외 객체 출력
            self.logger.error(f"{url}: Exception occurred: {e}")
            # traceback.print_exc()


        if total_word_count > 0:
            self.redis_client.set(url, total_word_count)
```
